```python
import pandas as pd
import numpy as np
import os
file_path = os.path.abspath(os.path.join('..', 'data', 'quikr_car.csv'))
data = pd.read_csv(file_path)
```

```python
data['Price']=data['Price'].str.replace(',','',regex=False)
data['Price']=pd.to_numeric(data['Price'], errors='coerce')
data = data.dropna(subset=['Price'])
data['Price']=data['Price'].astype(int)
data.info()
```

```python
data['kms_driven']=data['kms_driven'].str.replace(',','',regex=False)
data['kms_driven']=data['kms_driven'].str.replace('kms','',regex=False)
data['kms_driven'] = pd.to_numeric(data['kms_driven'], errors='coerce')
data = data.dropna(subset=['kms_driven'])
data['kms_driven'] = data['kms_driven'].astype(int)
data.info()
data.head()
```

```python
data=data.dropna()
data.info()
```

```python
data['name']=data['name'].str.split(' ').str.slice(0,3).str.join(' ')
data.head()
```

```python
data=data.reset_index(drop=True)
data.head()
```
